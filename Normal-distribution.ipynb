```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic=pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
from scipy.stats import norm
```

```python
sns.kdeplot(titanic["Age"])
```

```python
mean=titanic['Age'].mean()
print(mean)
```

```python
sd=titanic["Age"].std()
print(sd)
```

```python
x=(titanic['Age']-mean)/sd
print(x)
```

```python
sns.kdeplot(x)
```

```python
x.mean()
```

```python
x.std()
```

```python
from numpy.random import normal
```

```python
sample=normal(loc=0,scale=1,size=10000);
```

```python
sample.mean()
```

```python
sample.std()
```

```python
sample=(sample-sample.mean())/sample.std()
```

```python
sns.kdeplot(sample)
```

```python
dist=norm(sample.mean(),sample.std())
values=np.linspace(sample.min(),sample.max(),10000)
```

```python
value2=[dist.pdf(value) for value in values]
```

```python
plt.plot(values,value2)
```
